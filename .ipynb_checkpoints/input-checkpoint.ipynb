##3.2 wrangling

```python
bill_price=transactions.totalprice.groupby([transactions.id,transactions.billno]).sum()
bill_Hprice=bill_price.groupby(bill_price.index.droplevel(1)).max()
bill_Hprice
```

```python
grouped=transactions.billno.groupby([transactions.id])
frequency=grouped.nunique()
frequency
```

```python
docCountperItem=transactions.doc.groupby([transactions.id,transactions.unitprice]).count()
difdoc=docCountperItem.groupby(docCountperItem.index.droplevel(1)).mean()
difdoc
```

```python
experience=(patients.age+20000*patients.retire)
```

```python
fig=plt.figure()
fig.add_subplot(111)
experience.hist(bins=100)
```

#4 get Sets

```python
wholeSet=\
pd.DataFrame(data=[bill_Hprice/(bill_Hprice.max()-bill_Hprice.min()),
                  frequency/(frequency.max()-frequency.min()),
                  difdoc/(difdoc.max()-difdoc.min())],
            index=['bill_Hprice',
                  'frequency',
                  'difdoc']).T
```

```python
experience
```

```python
wholeSet=pd.merge(pd.DataFrame([patients.isCared,patients.id,experience/(experience.max()-experience.min())],
                               index=['isCared','id','experience']).T,
                  wholeSet,
                  left_on='id',
                  right_index=True,
                  how='inner')
```

```python
wholeSet=wholeSet.dropna()
```

```python
stats=wholeSet.describe()
print stats
stats.frequency['max']
```

```python
wholeSet.experience.hist(bins=100)
plt.show()
```

```python
transactions.describe()
```

```python
defineTrain=(wholeSet.isCared==0)|(transactions.totalCost<500) #|(wholeSet.experience<0.1)
defineTrain
```

```python
trainSet_ref=wholeSet[defineTrain].reset_index().drop(['isCared','index'],axis=1)
trainSet=np.array(trainSet_ref.drop('id',axis=1))
trainSet
```

```python
testSet_ref=wholeSet[~defineTrain].reset_index().drop(['isCared','index'],axis=1)

testSet=np.array(testSet_ref.drop('id',axis=1))
testSet
```

```python
testSet.T[3]*=0.1
```

#5 OCSVM

```python
from sklearn.covariance import EllipticEnvelope
from sklearn.svm import OneClassSVM
import matplotlib.font_manager
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define "classifiers" to be used
classifiers = {
    "OCSVM": OneClassSVM(nu=0.001,kernel='rbf', gamma=0.1)}
```

```python
clf=classifiers['OCSVM']
clf.fit(trainSet)
```

```python
Z = clf.decision_function(np.c_[testSet.T[0],
                                testSet.T[1],
                                testSet.T[2],
                                testSet.T[3]])
```

```python
xyz1=lambda x: testSet.T[x][Z.T[0]>0][:500]
xyz0=lambda x: testSet.T[x][Z.T[0]<0][:200]
```
